analyzing the data and getting some graphs

```python
import matplotlib.pyplot as plt
value_counts = df['Do you believe that having a project positively impacted your overall performance in the course? '].value_counts()
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%')
plt.title('Do you believe that having a project positively impacted your overall performance in the course? ')
```

```python
plt.hist(df['How would you rate your understanding of the course material? '])
plt.xlabel('students level of understanding of their course material that included a project')
plt.ylabel('frequency')
plt.title('student level of understanding')
```

```python
value_counts = df['What was your grade in the course in the last semester? '].value_counts()
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%')
plt.title('What was your grade in the course in the last semester? ')
```

```python
for index, value in df['Do you believe that having a project positively impacted your overall performance in the course? '].items():
  if value=='Yes':
    df.loc[index,'Do you believe that having a project positively impacted your overall performance in the course? '] = 1
  else:
    df.loc[index,'Do you believe that having a project positively impacted your overall performance in the course? '] = 0
df.head(19)
```

```python
# @title Age:

from matplotlib import pyplot as plt
df['Age:'].plot(kind='hist', bins=20, title='Age:')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Gender:

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Gender:').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title What was your grade in the course in the last semester?  vs Do you believe that having a project positively impacted your overall performance in the course?

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Do you believe that having a project positively impacted your overall performance in the course? '].value_counts()
    for x_label, grp in df.groupby('What was your grade in the course in the last semester? ')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('What was your grade in the course in the last semester? ')
_ = plt.ylabel('Do you believe that having a project positively impacted your overall performance in the course? ')
```

```python
# @title What was your grade in the course in the last semester?

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('What was your grade in the course in the last semester? ').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

some statictics that concerns the dataset

```python
# Calculate the mean
mean = df.mean()

# Calculate the median
median = df.median()

# Calculate the mode
mode = df.mode().iloc[0]

# Calculate the standard deviation
std_dev = df.std()

# Calculate the minimum
minimum = df.min()

# Calculate the maximum
maximum = df.max()

# Print the calculated statistics
print("Mean:")
print(mean)
print("\nMedian:")
print(median)
print("\nMode:")
print(mode)
print("\nStandard Deviation:")
print(std_dev)
print("\nMinimum:")
print(minimum)
print("\nMaximum:")
print(maximum)
```
